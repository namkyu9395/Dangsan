```python
def insta_searching(word):
    url = "https://www.instagram.com/explore/tags/" + word
    return url
```

```python
def insta_login(email,password):
    input_id = driver.find_elements_by_css_selector('input._2hvTZ.pexuQ.zyHYP')[0]
    input_id.clear()
    input_id.send_keys(email)
    
    input_pw = driver.find_elements_by_css_selector('input._2hvTZ.pexuQ.zyHYP')[1]
    input_pw.clear()
    input_pw.send_keys(password)
    input_pw.submit()
    time.sleep(3)
    
    save_info = driver.find_element_by_css_selector('#react-root > section > main > div > div > div > section > div > button')
    save_info.click()
    time.sleep(3)
    
    alarm = driver.find_element_by_css_selector('body > div.RnEpo.Yx5HN > div > div > div > div.mt3GC > button.aOOlW.HoLwm')
    alarm.click()
```

#### ì²«ë²ˆì§¸ ê²Œì‹œê¸€ ì—´ê¸°

```python
def select_first(driver):
    
    # find_element_by_css_selector : BeutifulSoupì„ ì‚¬ìš©í•˜ì§€ ì•Šê³  Seleniumì—ì„œ ì§ì ‘ htmlì— ì ‘ê·¼í•˜ëŠ” ë©”ì˜ë“œ
    first = driver.find_element_by_css_selector('div._9AhH0')
    
    # ê²Œì‹œë¬¼ í´ë¦­
    first.click()
    
    # ì •ë³´ ë¡œë”©ì— ì‹œê°„ì´ í•„ìš”í•˜ë¯€ë¡œ 3ì´ˆê°„ ëŒ€ê¸°
    time.sleep(3)
```

#### ê²Œì‹œê¸€ ì •ë³´ ê°€ì ¸ì˜¤ê¸°

```python
# ì •ê·œ í‘œí˜„ì‹ì„ ìœ„í•œ í´ë˜ìŠ¤
import re

def get_content(driver):
    # 1. í˜„ì¬ í˜ì´ì§€ì˜ HTMLì •ë³´ ê°€ì ¸ì˜¤ê¸°
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    
    # 2. ë³¸ë¬¸ ë‚´ìš© ê°€ì ¸ì˜¤ê¸°
    try:
        content = soup.select('div.C4VMK > span')[0].text
    except:
        content = ''
        
    # 3. ë³¸ë¬¸ ë‚´ìš©ì—ì„œ í•´ì‹œíƒœê·¸ ê°€ì ¸ì˜¤ê¸°(ì •ê·œí‘œí˜„ì‹ í™œìš©)
    tags = re.findall(r'#[^\s#,\\]+', content)
    
    # 4. ì‘ì„±ì¼ì ì •ë³´ ê°€ì ¸ì˜¤ê¸°
    date = soup.select('time.FH9sR.Nzb55')[0]['datetime'][:10]
    
    # 5. ì¢‹ì•„ìš” ìˆ˜ ê°€ì ¸ì˜¤ê¸°
    try:
        like = soup.select('div.Nm9Fw > button')[0].text[4:-1]
    except:
        like = 0
        
    # 6. ìœ„ì¹˜ì •ë³´ ê°€ì ¸ì˜¤ê¸°
    try:
        place = soup.select('div.JF9hh > a.O4GlU')[0].text
    except:
        place = ''
    
    # 7. ìˆ˜ì§‘í•œ ì •ë³´ ì €ì¥í•˜ê¸°
    data = [content, date, like, place, tags]
    return data
```

```python
# ë‹¤ìŒ ê²Œì‹œê¸€ì„ ì—¬ëŠ” í•¨ìˆ˜

def move_next(driver):
    right = driver.find_element_by_css_selector('a._65Bje.coreSpriteRightPaginationArrow')
    right.click()
    time.sleep(3)
```

### í¬ë¡¤ë§ ê³¼ì • í†µí•© í•¨ìˆ˜

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import re

# 1. í¬ë¡¬ë¸Œë¼ìš°ì € ì—´ê¸°
driver = webdriver.Chrome('C:/Users/NK/Desktop/programming/python/chromedriver.exe')

# 2. ì¸ìŠ¤íƒ€ê·¸ë¨ì— ì ‘ì†
driver.get('https://www.instagram.com')
time.sleep(2)

# 3. ì¸ìŠ¤íƒ€ê·¸ë¨ ë¡œê·¸ì¸
email = '***********'
pw = '*************'
insta_login(email,pw)

# 4. ì¸ìŠ¤íƒ€ê·¸ë¨ ê²€ìƒ‰í˜ì´ì§€ URL ë§Œë“¤ê¸°
word = 'ì œì£¼ì—¬í–‰' # ê²€ìƒ‰ì–´
url = insta_searching(word)

# 5. ê²€ìƒ‰í˜ì´ì§€ ì ‘ì†í•˜ê¸°
driver.get(url)
time.sleep(3)

# 6. ì²« ë²ˆì§¸ ê²Œì‹œê¸€ ì—´ê¸°
select_first(driver)

# 7. ë¹„ì–´ìˆëŠ” ë³€ìˆ˜(results) ë§Œë“¤ê¸°
results = []

# 8. ì—¬ëŸ¬ ê²Œì‹œë¬¼ ìˆ˜ì§‘í•˜ê¸°
target = 100
for i in range(target):
    data = get_content(driver)
    results.append(data)
    move_next(driver)
    
print(results[:2])
```

### ìˆ˜ì§‘ ë°ì´í„° ì €ì¥

```python
import pandas as pd

results_df = pd.DataFrame(results)
results_df.columns = ['content', 'data', 'like', 'place', 'tags']
results_df.to_excel('crawling_jejutravel.xlsx')
```
