### 카카오 API를 활용한 장소 검색

```python
# 장소 검색 예시

import requests

searching = '합정 스타벅스'
url = f'https://dapi.kakao.com/v2/local/search/keyword.json?query={searching}'
headers = {'Authorization' : 'KakaoAK **************************'}
places = requests.get(url, headers = headers).json()['documents']
places
```

```python
# 카카오 API로 장소 검색 함수 만들기

def find_places(searching):

    # 1. 접속 url 만들기
    url = f'https://dapi.kakao.com/v2/local/search/keyword.json?query={searching}'

    # 2. headers 입력하기
    headers = {'Authorization' : 'KakaoAK ***********************'}
    
    # 3. API 요청&정보 받기
    places = requests.get(url, headers = headers).json()['documents']
    
    # 4. 필요한 정보 선택하기
    place = places[0]
    name = place['place_name']
    x = place['x']
    y = place['y']
    data = [name, x, y, searching]
    
    return data
```

```python
# 제주공항 검색 예시

data = find_places('제주공항')
data
```

```python
# 인스타그램 위치명에 대한 위치 정보 검생


## 작업 진행현황을 볼 수 있는 클래스
from tqdm import tqdm_notebook
import time

locations_inform = []
for location in tqdm_notebook(locations):
    try:
        data = find_places(location)
        locations_inform.append(data)
        time.asleep(0.5)
    except:
        pass
locations_inform
```

```python
locations_inform_df = pd.DataFrame(locations_inform)
locations_inform_df.columns = ['name_official', '경도', '위도', '인스타위치명']
locations_inform_df.to_excel('crawling_data/locations.xlsx', index=False)
```

```python
# 인스타 게시량 및 위치정보 데이터 불러오기
locations_count_df = pd.read_excel('crawling_data/location_counts.xlsx',index_col = 0)
locations_inform_df = pd.read_excel('crawling_data/locations.xlsx')
```

```python
# 위치 데이터 병합

## 왼쪽 데이터의 name_official칼럼과 오른쪽 데이터의 인덱스를 기준으로 병합할 경우
## left_on = '칼럼명', right_index=True
location_data = pd.merge(locations_inform_df, locations_count_df,
                         how = 'inner', left_on = 'name_official',right_index=True)
location_data.head()
```

```python
# 병합한 데이터 저장

location_data.to_excel('crawling_data/location_inform.xlsx', index = False)
```

### folium을 이용한 지도 시각화 (1) 개별표시

```python
location_data = pd.read_excel('crawling_data/location_inform.xlsx')
location_data.info()
```

```python
import folium

## 중심이 되는 기준점의 좌표
Mt_Hanla = [33.362500, 126.533694]
map_jeju = folium.Map(location = Mt_Hanla, zoom_start=11)

for i in range(len(location_data)):
    name = location_data['name_official'][i]  # 공식명칭
    count = location_data['place'][i]         # 게시글 개수
    size = int(count) * 2                     # 원의 크기
    long = float(location_data['위도'][i])
    lat = float(location_data['경도'][i])
    ## 위도 경도 데이터를 통해 원을 생성하고 map_jeju에 추가할 것
    folium.CircleMarker((long, lat), radius = size, color = 'red', popup = name).add_to(map_jeju)
    
map_jeju
```

```python
# 지도 저장
map_jeju.save('crawling_data/jeju.html')
```

### folium을 이용한 지도 시각화 (2) 그룹으로 표시

```python
# 서클마커가 특정 지역에 몰려있을 경우 그룹으로 묶어 알아보기 쉽게 시각화

from folium.plugins import MarkerCluster

locations = []
names = []

for i in range(len(location_data)):
    data = location_data.iloc[i] #행 하나씩
    locations.append((float(data['위도']), float(data['경도'])))
    names.append(data['name_official'])
    
icon_create_function = '''\
function(cluster) {
    return L.divIcon({
    html : '<b>' + cluster.getChildCount() + '</b>',
    className : 'marker-cluster marker-cluster-large',
    iconSize : new L.Point(30,30)
    });
}'''

Mt_Hanla = [33.362500, 126.533694]
map_jeju2 = folium.Map(location = Mt_Hanla, zoom_start = 11)

marker_cluster = MarkerCluster(locations = locations, popups = names, name = 'Jeju',
                               overlay = True, control = True, icon_create_function = icon_create_function)

marker_cluster.add_to(map_jeju2)
folium.LayerControl().add_to(map_jeju2)

map_jeju2
```

```python
# 지도 저장

map_jeju2.save('crawling_data/jeju_cluster.html')
```
