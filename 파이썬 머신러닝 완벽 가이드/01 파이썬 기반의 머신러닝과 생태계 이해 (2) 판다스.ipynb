### 2. DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환

```python
# 리스트와 넘파이 ndarray로 1차원 데이터프레임 만들기

import numpy as np

col_name1 = ['col1']
list1 = [1,2,3]
array1 = np.array(list1)
print('array1 shape : ', array1.shape)

# 리스트를 이용하여 dfm 생성
df_list1 = pd.DataFrame(list1, columns = col_name1)
print('1차원 리스트로 만든 DataFrame:\n', df_list1)

# 넘파이 ndarray를 이용하여 dfm 생성
df_array1 = pd.DataFrame(array1, columns = col_name1)
print('1차원 ndarray로 만든 DataFrame:\n', df_array1)
```

```python
# 리스트와 넘파이 ndarray로 2차원 데이터 프레임 만들기
## 데이터프레임은 2차까지 밖에 없음!

## 2행 3열 형태의 리스트로 2개의 인덱스, 3개의 칼럼을 가진 데이터 프레임 생성
list2 = [[1,2,3],
         [4,5,6]]
array2 = np.array(list2)
print('array2 shape:', array2.shape)
## 3열이므로 3개의 칼럼명 필요
col_list2 = ['col1', 'col2', 'col3']

df_list2 = pd.DataFrame(list2, columns = col_list2)
print('2차원 리스트로 만든 dfm : \n', df_list2)
df_array2 = pd.DataFrame(array2, columns = col_list2)
print('2차원 ndarray로 만든 dfm : \n', df_array2)
```

```python
# 딕셔너리로 데이터프레임 만들기

## 딕셔너리의 Key는 문자열 칼럼명으로 매핑, Value는 칼럼데이터로 매핑
dict = {'col1' : [1,2,3], 
        'col2' : [4,5,6],
        'col3' : [7,8,9]}
df_dict = pd.DataFrame(dict)
print('딕셔너리로 만든 dfm : \n', df_dict)
```

```python
# dfm을 넘파이 ndarray, 리스트, 딕셔너리로 변환하기

# ndarray로 변환하기
array3 = df_dict.values
print('df_dict.values 타입 : ', type(array3), 'df_dict.values shape : ', array3.shape)
print(array3)

# 리스트로 변환하기
list3 = df_dict.values.tolist()
print('\ndf_dict.values.tolist() 타입 :', type(list3))
print(list3)

# 딕셔너리로 변환
dict3 = df_dict.to_dict('list')
print('\n df_dict.to_dict() 타입:', type(dict3))
print(dict3)
```

### 8. apply lambda 식으로 데이터 가공

```python
# def로 함수를 만들 수 있음
def get_square(a):
    return a**2

print('3의 제곱은 ', get_square(3))
```

```python
# labmda로 함수를 간단하게 표현할 수 있음
lambda_square = lambda x : x**2
print('3의 제곱은 ', lambda_square(3))
```

```python
# lambda식으로 여러개의 값을 입력 인자로 사용할 경우 보통 map() 함수 활용

a = [1,2,3]
squares = map(lambda x : x**2,a)
list(squares)
```

```python
# 판다스에서는 apply메서드를 이요하여 lambda식 사용

titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name', 'Name_len']].head()
```

```python
# if else절을 결합하여 복잡한 lambda 식 만들기

## 콜론(:) 다음이 반환값
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x <= 15 else 'Adult')
titanic_df[['Age', 'Child_Adult']].head(8)
```

```python
# lambda는 elif를 지원하지 않으므로 3이상의 카테고리는 괄호를 사용하여 다시 if else를 적용해주어야함
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else ('Adult' if x <= 60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
# 조건이 많다면 lambda식보단 별도의 함수를 만드는 것이 가독성이 높음

def get_category(age):
    cat = ''
    if age <= 5: cat = 'Baby'
    elif age <= 12: cat = 'Child' 
    elif age <= 18: cat = 'Teenager'
    elif age <= 25: cat = 'Student'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 60: cat = 'Adult'
    else : cat = 'Eldely'
        
    return cat

titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age', 'Age_cat']].head(10)
```
