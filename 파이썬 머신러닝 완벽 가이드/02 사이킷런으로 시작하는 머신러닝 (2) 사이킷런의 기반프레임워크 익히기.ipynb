|API명|설명|
|:------|:---|
|datasets.make_classifications()|분류를 위한 데이터 세트 생성|
|datasets.make_blobs()|클러스터링을 위한 데이터 세트 생성|



    ** 대다수의 데이터세트가 딕셔너리의 형태로 구성
        Key값 - 1) data : 피처의 데이터 세트
                2) target : 분류일때는 레이블 값, 회귀일때는 숫자 결괏값 데이터 세트
                3) target_names : 개별 레이블의 이름
                4) feature_names : 피처의 이름
                5) DESCR : 데이터 세트에 대한 설명과 피처의 설명

```python
# 붓꽃 데이터 세트 생성
from sklearn.datasets import load_iris

iris_data = load_iris()
print(type(iris_data))
```

```python
# 딕셔너리 형태(Bunch 클래스)인 붓꽃 데이터의 key값 확인
keys = iris_data.keys()
print('붓꽃 데이터 세트의 키들 : ', keys)
```

```python
# Key들이 가리키는 값 확인

print('feature_names의 type:', type(iris_data.feature_names))
print('feature_names의 shape:', len(iris_data.feature_names))
print(iris_data.feature_names)

print('\n target_names의 type:', type(iris_data.target_names))
print('target_names의 shape:', len(iris_data.target_names))
print(iris_data.target_names)

print('\n data의 type:', type(iris_data.data))
print('data의 shape:', iris_data.data.shape)
print(iris_data['data'])

print('\n target의 type:', type(iris_data.target))
print('target의 shape:', iris_data.target.shape)
print(iris_data.target)
```
